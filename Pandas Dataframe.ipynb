# Dropping columns

```python
import pandas as pd
df_iris = pd.read_csv('Iris.csv')
df_iris
```

```python
# axis=1 indicates all of the cols
df_iris.drop('SepalLengthCm', axis=1)
```

```python
# axis=0 indicates rows
df_iris.drop([3,4,5], axis=0)
```

```python
# display names of columns in the dataframe
df_iris.columns
```
